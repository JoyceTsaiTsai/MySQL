# 4.巴菲特說：挑選投資的公司時，ROE必須達到15%，
#但是我不喜歡紡纖股，
#請幫我挑出ROE>0.15的【非紡纖股】股票，請排序(高到低)
#ROE = 股東權益報酬率
資料表：stock2022

```python
sql = "\
select 代碼,名稱,股東權益報酬率,類股 \
from stock2022 \
where 類股 !='紡纖股' and 股東權益報酬率>=0.15 \
order by 股東權益報酬率 desc;"
df = pd.read_sql_query(sql, conn)
df
```

# 7.本公司想要培養儲備幹部，請找出被評語有『認真』，且年紀在25歲以下的
資料表：員工


```python
sql = "\
select 姓名,稱呼,職稱,(2023-year(出生日期)+1) as 年紀,附註 \
from 員工 \
where (2023-year(出生日期)+1)<=25 and 附註 like '%認真%';"
df = pd.read_sql_query(sql, conn)
df
```

# 9.顯示本公司員工生日在monday的壽星資料
資料表：員工

```python
sql = "\
select 姓名,稱呼,職稱,date_format(出生日期,'%W') 出生星期 \
from 員工 \
where date_format(出生日期,'%W')='Monday';"
df = pd.read_sql_query(sql, conn)
df
```

# 10.顯示本公司員工年齡最大的員工資料

# 11.顯示員工詳細資料，住址
資料表：員工

```python
sql = "\
select 姓名,replace(replace(稱呼,'先生', '男'),'小姐', '女') AS 性別, \
  concat(區域號碼,城市,行政區,地址) as 地址, \
  電話號碼 from 員工;"
df = pd.read_sql_query(sql, conn)
df
```

# 12.顯示員工的『姓，名』
資料表：員工

```python
sql = "\
select substring(姓名,1,1) as 姓, \
  substring(姓名,2,2) as 名, \
  concat(substring(姓名,2,2),稱呼) as title from 員工;"
df = pd.read_sql_query(sql, conn)
df
```
