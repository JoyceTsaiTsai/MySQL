#world資料庫的3個資料表，說明
MySQL的世界資料庫是一個用於存儲全球城市、國家和國家語言資訊的資料庫。以下是這三個資料表的欄位說明：
#1.	city 資料表：
**o	ID：城市的唯一識別碼。**

o	Name：城市名稱。

o	CountryCode：城市所在國家的ISO 3166-1代碼，用於關聯到country資料表。

o	District：城市所在地區的名稱。

o	Population：城市的人口數量。

---
#2.	country 資料表：
**o	Code：國家的ISO 3166-1代碼**

o	Name：國家名稱。

o	Area：國家面積（以平方公里為單位）。

o	IndepYear：國家獨立年份。

o	Population：國家人口數量。

o	LifeExpectancy：國家的人均預期壽命。

o	LocalName：國家的地方名稱（非英語）。

o	GovernmentForm：國家的政府形式。

o	HeadOfState：國家元首的姓名。

**o	Capital：國家的首都城市（通過外鍵關聯到city資料表的ID欄位）。**

---
#3.	countrylanguage 資料表：
**o	CountryCode：國家的ISO 3166-1代碼，用於關聯到country資料表。**

o	Language：國家使用的語言。

o	IsOfficial：該語言是否為官方語言。

o	Percentage：使用該語言的的人口比例（根據該語言的母語者計算）。

# 1.顯示『城市的人口超過700000人，而且語言是德語Dutch的城市資料』


資料庫：world

資料表：

city：城市

countrylanguage：語言

```python
import pandas as pd
sql = "select Name as 城市,Language as 語言,Population as 人口 from city \
join countrylanguage \
	on city.CountryCode=countrylanguage.CountryCode \
where Population>700000 and Language='Dutch';"
df = pd.read_sql_query(sql, conn)
df
```

# 2.顯示以下國家『'Austria','China','Spain','France','Greece','Iran','United States'』的首都
資料庫：world

資料表：

city：城市

country：國家

```python
import pandas as pd
sql = "select country.Name as 國家,city.Name as 首都 from city \
join country \
	on city.CountryCode=country.`Code` \
where country.Name in ('Austria','China','Spain','France','Greece','Iran','United States') and ID=capital;"
df = pd.read_sql_query(sql, conn)
df
```

#6.台灣Taiwan的城市，依照人口數排序（高到低），最後顯示排行第2，第4的城市資料
資料庫：world

資料表：

city：城市

country：國家

```python
import pandas as pd
sql = "select * \
from \
(select city.Name as 城市,country.Name as 國名,city.Population as 城市人口 from city \
join country \
	on city.CountryCode=country.`Code` \
where country.Name='Taiwan' order by city.Population desc limit 1,1) as t1 \
union \
select * \
from \
(select city.Name as 城市,country.Name as 國名,city.Population as 城市人口 from city \
join country \
	on city.CountryCode=country.`Code` \
where country.Name='Taiwan' order by city.Population desc limit 3,1) as t2;"
df = pd.read_sql_query(sql, conn)
df
```

# 9.計算經管3a的學科平均，及其與平均的偏差值

資料表：經管3a

```python
import pandas as pd
sql = "select *,(平均-全班平均) 偏差值 \
from \
(select *, \
(select round(avg(平均),1) from (select *, \
round((中文+英文+數學)/3,1) 平均 \
from 經管3a) as t1)全班平均 \
from \
(select *, \
round((中文+英文+數學)/3,1) 平均 \
from 經管3a) as t1) as t2;"
df = pd.read_sql_query(sql, conn)
df
```

# 10.顯示經管3a同學的學科平均，註明不及格，標註評等，標註第1名/最後1名
A：90(含)以上

B：80(含)～90

C：60(含)～80

D：60以下

資料庫：firstdb

資料表：經管3a

```python
#名次:先找出最高與最低分->'珍妮弗'、'約翰'
import pandas as pd
sql = "select *, \
case \
	when 姓名='約翰' then '最後1名' \
    when 姓名='珍妮弗' then '第1名' \
    else '' \
end as 名次 \
from \
(select *, \
((中文+英文+數學)/3) 平均, \
case \
	when ((中文+英文+數學)/3)<60 then '不及格' \
	else '' \
end as 不及格, \
case \
	when ((中文+英文+數學)/3)>=90 then 'A' \
    when ((中文+英文+數學)/3) between 80 and 90 then 'B' \
    when ((中文+英文+數學)/3) between 60 and 80 then 'C' \
	else 'D' \
end as 評等 \
from 經管3a order by ((中文+英文+數學)/3)) as t1 order by 學號;"
df = pd.read_sql_query(sql, conn)
df
```
