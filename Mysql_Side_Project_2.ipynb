# 1.顯示電影的『評等』為『好評』的電影
#評等標準：
#rating>=9為好評，

#8.3<=rating<9為普通，

#rating<8.3為差評


資料庫：imdb

casting：電影選角

movies：電影

actors：演員

```python
import pandas as pd
sql = " \
select title,release_year, \
  case \
		when rating>=9 then '好評' \
		when rating<9 and rating>=8.3 then '普通' \
		else '差評' \
	end as 評等 \
from movies \
where rating>=9;"
df = pd.read_sql_query(sql, conn)
df
```

#6.顯示本班，男女的人數，百分比
資料庫：firstdb

資料表：經管2a

```python
import pandas as pd
sql = " \
select count(*) as 總人數, \
(select count(*) from 經管2a where 性別='男') as 男生人數, \
(select count(*) from 經管2a where 性別='女') as 女生人數, \
(select count(*) from 經管2a where 性別='男')/count(*) as 男生比例, \
(select count(*) from 經管2a where 性別='女')/count(*) as 女生比例 \
from 經管2a;"
df = pd.read_sql_query(sql, conn)
df
```

# 8.顯示問卷調查不同教育程度的人數
資料庫：firstdb

資料表：問卷

```python
import pandas as pd
sql = " \
select (SELECT COUNT(*) from 問卷 where 教育程度=1)/count(*) as 高中比例, \
(SELECT COUNT(*) from 問卷 where 教育程度=2)/count(*) as 專科比例, \
(SELECT COUNT(*) from 問卷 where 教育程度=3)/count(*) as 大學比例, \
(SELECT COUNT(*) from 問卷 where 教育程度=4)/count(*) as 研究所比例 \
from 問卷;"
df = pd.read_sql_query(sql, conn)
df
```

```python
import pandas as pd
sql = "select 教育程度,count(*)/(SELECT COUNT(*) from 問卷) as 教育程度比例 from 問卷 group by 教育程度;"
df = pd.read_sql_query(sql, conn)
df
```

# 9.顯示員工『郭逸洋』總共負責那些書籍
資料庫：firstdb

資料表：books, bookemployee, bookorder

```python
import pandas as pd
sql = "select \
書籍名稱,姓名 as 負責員工姓名 \
from books \
join bookemployee \
	on books.負責員工編號=bookemployee.員工編號 \
where 姓名='郭逸洋';"
df = pd.read_sql_query(sql, conn)
df
```
